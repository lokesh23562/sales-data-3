```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
sales = pd.read_csv("/content/sample_data/sales_data (3).csv")
customers = pd.read_csv("/content/sample_data/customer_churn.csv")

# Date + calculated column
# Fixed: Using 'Date', 'Quantity', and 'Price' to match the CSV columns
sales['order_date'] = pd.to_datetime(sales['Date'])
sales['total_amount'] = sales['Quantity'] * sales['Price']
sales['month'] = sales['order_date'].dt.month

# Merge datasets
# Fixed: Using 'Customer_ID' for sales and 'CustomerID' for customers
df = pd.merge(sales, customers, left_on='Customer_ID', right_on='CustomerID', how='left')

# --- AGGREGATIONS ---
monthly_sales = df.groupby('month')['total_amount'].sum()
# Note: Assuming customer_name/region might need specific column names from the churn file if not in sales
region_sales = df.groupby('Region')['total_amount'].sum()

# --- FILTERING ---
high_value = df[(df.total_amount > 1000) & (df.Region == 'South')]

# --- PIVOT TABLE ---
pivot = pd.pivot_table(df, values='total_amount', index='Region', columns='Product', aggfunc='sum')

# --- VISUALS ---
monthly_sales.plot(title="Monthly Sales"); plt.show()
region_sales.plot(kind='bar', title="Region Sales"); plt.show()

# --- KEY METRICS ---
print("Total Revenue:", df.total_amount.sum())
print("Total Customers:", df.Customer_ID.nunique())
```

```python
total_revenue = df['total_amount'].sum()
total_customers = df['Customer_ID'].nunique()
average_order_value = df['total_amount'].mean()

print("Total Revenue:", total_revenue)
print("Total Customers:", total_customers)
print("Average Order Value:", average_order_value)
```
